```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import xarray as xr
import scipy.special
```

```python
tau_0 = 20
delta_omega = 2/tau_0
tau_t = 1.0*tau_0

z = 0
x = np.linspace(-4*np.sqrt(1+z**2), 4*np.sqrt(1+z**2), 201)
y = np.linspace(-4*np.sqrt(1+z**2), 4*np.sqrt(1+z**2), 101)
time = np.linspace(-5*tau_0, +5*tau_0, 1000)

E_time = xr.DataArray(np.zeros((len(y), len(x), len(time)), dtype=complex),
                      coords=[y, x, time],
                      dims=['y', 'x', 'time'])
                      
psi_G = np.arctan(z)
b = tau_t/2
qz = z-1j
sz = np.sqrt(1+z**2)
Amp = np.exp(1j*E_time.coords['y']**2/(qz))*np.exp(1j*E_time.coords['x']**2/(qz))*np.exp(-1j*E_time.coords['time'])*np.exp(-1j*2*psi_G)
alpha = np.sqrt((1/delta_omega)**2 - 1j*b**2/qz)
tp = E_time.coords['time']
tpp = tp + 2*b*E_time.coords['x']/(qz) - (E_time.coords['x']**2+E_time.coords['y']**2)/(qz)

E_x_an = Amp*np.exp(-(tpp/(2*alpha))**2)*(E_time.coords['x'] + 1j*b*tpp/(2*alpha**2))/(alpha*delta_omega)

E_y_an = Amp*np.exp(-(tpp/(2*alpha))**2)*E_time.coords['y']/(alpha*delta_omega)

E_z_an = 1j*Amp*np.exp(-(tpp/(2*alpha))**2)*(1 - (np.exp(-1j*2*psi_G)/sz)*(E_time.coords['x']**2 + E_time.coords['y']**2 
                                             + b**2/(2*alpha**2) + 1j*b*tpp*E_time.coords['x']/alpha**2
                                             - b**2*tpp**2/(4*alpha**4)))/(alpha*delta_omega)

plt.figure(figsize=[10, 7])
plt.subplot(2, 2, 1)
np.abs(E_x_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 2)
np.abs(E_x_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 3)
np.real(E_x_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 4)
np.real(E_x_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.tight_layout()
```

```python
plt.figure(figsize=[10, 7])
plt.subplot(2, 2, 1)
np.abs(E_y_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 2)
np.abs(E_y_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 3)
np.real(E_y_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 4)
np.real(E_y_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.tight_layout()
```

```python
plt.figure(figsize=[10, 7])
plt.subplot(2, 2, 1)
np.abs(E_z_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 2)
np.abs(E_z_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 3)
np.real(E_z_an).sel(x=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.subplot(2, 2, 4)
np.real(E_z_an).sel(y=0).plot()
plt.xlim(-3*tau_0, 3*tau_0)
plt.tight_layout()
```

```python
# Define figure dimensions
fig_width = 10/2.54  # width in inches
fig_height = 3/2.54       # height in inches
fig_size = [fig_width,fig_height]
# tell matplotlib to use latex font rendering
plt.rc('text', usetex=False)
# adjust fontsize
plt.rcParams.update({'font.size' : 6,
          'axes.labelsize' : 6,
          'legend.fontsize': 6,
          'xtick.labelsize' : 6,
          'ytick.labelsize' : 6,
          'lines.linewidth'   : 0.75,
          'lines.markersize'   : 2})
# adjust resolution
plt.rcParams['savefig.dpi'] = 600

fig=plt.figure(figsize=fig_size)

plt.subplot(1, 2, 1)
np.real(E_z_an).sel(y=0, method='nearest').plot(add_colorbar=False)
plt.xlim(-4*tau_0, 4*tau_0)
plt.ylim(-2.5, 2.5)
plt.xlabel(r'$\omega_0 t$')
plt.ylabel(r'$x/w_0$')
plt.title('')
plt.subplot(1, 2, 2)
np.real(E_x_an).sel(y=0, method='nearest').plot(add_colorbar=False)
plt.xlim(-4*tau_0, 4*tau_0)
plt.ylim(-2.5, 2.5)
plt.xlabel(r'$\omega_0 t$')
plt.ylabel(r'$x/w_0$')
plt.title('')
plt.tight_layout()
#plt.savefig('RPLB_SCanalytical.png', facecolor='white')
```
