```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import xarray as xr
import h5py
from scipy.special import gamma
from tqdm import tqdm
```

```python
# initialize constants (SI units)
c = 2.99792458e8 #speed of light
e_0 = 8.85418782e-12
# set Porras factor
g_0 = 'constant'
# set pure spectral properties
lambda_0 = 800e-9  # central wavelength
#quick frequency properties
omega_0 = 2*np.pi*c/lambda_0
s = 35
tau_0 = s*np.sqrt(np.exp(2/(s+1))-1)/omega_0
print(tau_0)
# calculate Rayleigh range and beam waist
w_00 = 4e-6
z_R0 = (omega_0*w_00**2)/(2*c)
```

```python
t_start = -10*tau_0
t_end = 10*tau_0
n = 20
num_t = np.int_(np.round(n*(t_end-t_start)/(lambda_0/c)))
omega = np.linspace(1,n*omega_0,num_t)
omega_step = omega[1]-omega[0]
time = np.fft.fftshift(np.fft.fftfreq(num_t,omega_step/(2*np.pi)))
dt = time[1]-time[0]

z = np.array([-5*z_R0, 0, 5*z_R0])
r = np.linspace(-5*w_00, 5*w_00, 801)

temp = xr.DataArray(np.zeros([len(z), len(r), len(omega)]),
                   coords=[z, r, omega],
                   dims=['z', 'r', 'omega'])

pulse_temp = np.sqrt(2*np.pi)*(s*temp.coords['omega']/omega_0)**(s+1)*(1/temp.coords['omega'])*np.exp(-s*temp.coords['omega']/omega_0)/gamma(s+1)

if (g_0 == 'constant'):
    w_0 = w_00
    z_R = z_R0
else:
    w_0 = w_00*(omega_0/temp.coords['omega'])**((g_0+1)/2)
    z_R = z_R0*(omega_0/temp.coords['omega'])**(g_0)

eps = w_0/z_R

fft_norm_coeff = np.sqrt(len(omega))*omega_step/np.sqrt(2*np.pi)

pulse_prep = pulse_temp*np.exp(1j*temp.coords['omega']*temp.coords['z']/c)
phi_G = np.arctan(temp.coords['z']/z_R)
w = w_0*np.sqrt(1+(temp.coords['z']/z_R)**2)
R = (temp.coords['z']) + (z_R**2)/(temp.coords['z'])
phi_norm = -(temp.coords['omega']/c)*(temp.coords['z']+(temp.coords['r']**2)/(2*R))
trans = np.exp(-(temp.coords['r']/w)**2)

c_2 = (w_0/w)**2 * np.exp(1j*(phi_norm + 2*phi_G))
c_3 = (w_0/w)**3 * np.exp(1j*(phi_norm + 3*phi_G))

rho = temp.coords['r']/w_0

E_z_spec = trans*pulse_prep*((c_2 - c_3*rho**2)*eps**2)
E_z_spec = E_z_spec.transpose('r', 'z', 'omega')
E_z_time = np.fft.fftshift(np.fft.ifft(E_z_spec, norm='ortho', axis=2), axes=2)*fft_norm_coeff
E_z_time = xr.DataArray(E_z_time,
                        coords=[r, z, time],
                        dims=['r', 'z', 'time'])

E_r_spec = trans*pulse_prep*((c_2*rho)*eps)
E_r_spec = E_r_spec.transpose('r', 'z', 'omega')
E_r_time = np.fft.fftshift(np.fft.ifft(E_r_spec, norm='ortho', axis=2), axes=2)*fft_norm_coeff
E_r_time = xr.DataArray(E_r_time,
                        coords=[r, z, time],
                        dims=['r', 'z', 'time'])
```

```python
fig = plt.figure(figsize=[10, 5])

ax1 = plt.subplot(1, 2, 1)
np.real(E_z_time).sel(z=0, method='nearest').plot()
plt.xlim(-2*tau_0, 2*tau_0)
plt.ylim(-4*w_00, 4*w_00)

ax2 = plt.subplot(1, 2, 2)
np.real(E_r_time).sel(z=0, method='nearest').plot()
plt.xlim(-2*tau_0, 2*tau_0)
plt.ylim(-4*w_00, 4*w_00)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=[10, 5])

ax1 = plt.subplot(1, 2, 1)
(np.abs(E_z_time)**2).sel(z=0, method='nearest').plot()
plt.xlim(-2*tau_0, 2*tau_0)
plt.ylim(-4*w_00, 4*w_00)

ax2 = plt.subplot(1, 2, 2)
(np.abs(E_r_time)**2).sel(z=0, method='nearest').plot()
plt.xlim(-2*tau_0, 2*tau_0)
plt.ylim(-4*w_00, 4*w_00)

plt.tight_layout()
```
